### Silver processing.

```python
df_bronze = spark.sql(f"SELECT * FROM {catalog}.{bronze_schema}.{data_source};")
df_bronze.show(10)
```

```python
df_bronze.printSchema()
```

```python
df_duplicates = df_bronze.groupBy("customer_id").count().where("count > 1")
display(df_duplicates)
```

```python
print('Rows before duplicate dropped: ', df_bronze.count())
df_silver = df_bronze.dropDuplicates(['customer_id'])
print('Rows after duplicate dropped: ', df_bronze.count())
```

```python
display(
    df_silver.filter(F.col("customer_name") != F.trim(F.col("customer_name")))
)
```

```python
df_silver=df_silver.withColumn("customer_name", F.trim(F.col("customer_name")))
```

```python
display(
    df_silver.filter(F.col("customer_name") != F.trim(F.col("customer_name")))
)
```

```python
df_silver.select('city').distinct().show()
```

```python
#typos -> correct names
city_mapping ={
    'Bengaluru': 'Bangalore',
    'Bengalore': 'Bangalore',

    'Hyderabadd': 'Hyderabad',
    'Hyderbad': 'Hyderabad',

    'NewDelhi': 'New Delhi',
    'NewDheli': 'New Delhi',
    'NewDelhee': 'New Delhi'

}

allowed = ['Bangalore', 'Hyderabad', 'New Delhi']

df_silver = (df_silver.replace(city_mapping, subset=["city"]).withColumn("city", F.when(F.col("city").isNull(),None).when(F.col("city").isin(allowed),F.col("city")).otherwise(None)))

df_silver.select('city').distinct().show()
```

```python
df_silver.select('customer_name').distinct().show()
```

```python
# Title case fix
df_silver=df_silver.withColumn("customer_name",F.when(F.col("customer_name").isNull(),None).otherwise(F.initcap(F.col("customer_name"))))

df_silver.select('customer_name').distinct().show()
```

```python
df_silver.filter(F.col("city").isNull()).show(truncate=False)
```

```python
null_customer_names = ['Sprintx Nutrition','Zenathlete Foods','Primefuel Nutrition','Recovery Lane']
df_silver.filter(F.col("customer_name").isin(null_customer_names)).show(truncate=False)
```

```python
# Business Confirmation Note: City corrections confirmed by business team
customer_city_fix = {
    # |Sprintx Nutrition
    789403:"New Delhi",
    # |Zenathlete Foods
    789420:"Bangalore",
    # |Primefuel Nutrition
    789520:"Bangalore",
    789521:"Hyderabad",
    # |Recovery Lane
    789603:"Hyderabad"
}

df_fix = spark.createDataFrame(
    [(k,v) for k,v in customer_city_fix.items()],
    ['customer_id','fixed_city']
)
display(df_fix)
```

```python
df_silver=(
    df_silver.join(df_fix, on='customer_id', how='left')
    .withColumn("city", F.coalesce("city","fixed_city"))     #Replace NUll city with fixed city
    .drop("fixed_city")
)
display(df_silver)
```

```python
df_silver=df_silver.withColumn("customer_id", F.col("customer_id").cast("string"))
print(df_silver.printSchema())
```

```python
df_silver = (
    df_silver
    # Build final customer column: "CustomerName-City" or "CustomerName-Unknown"
    .withColumn("customer",F.concat_ws("-","customer_name", F.coalesce(F.col("city"), F.lit("Unknown"))))

    # Static attributes aligned with parent data model
    .withColumn("market", F.lit("India"))
    .withColumn("platform", F.lit("Sports Bar"))
    .withColumn("channel", F.lit("Acquisition")))

display(df_silver.limit(10))
```

```python
df_silver.write\
    .format("delta")\
    .option("delta.enableChangeDataFeed", "true")\
    .option("mergeSchema", "true")\
    .mode("overwrite")\
    .saveAsTable(f"{catalog}.{silver_schema}.{data_source}")
```

### Gold Processing

```python
df_silver = spark.sql(f"SELECT * FROM {catalog}.{silver_schema}.{data_source};")

#Take req cols only
#customer_id, customer_name, city, read_timestamp, file_name, file_size, customer, market, platform, channel.

df_gold= df_silver.select("customer_id", "customer_name", "city", "customer", "market", "platform", "channel")

display(df_gold)
```

```python
df_gold.write\
    .format("delta")\
    .option("delta.enableChangeDataFeed", "true")\
    .mode("overwrite")\
    .saveAsTable(f"{catalog}.{gold_schema}.sb_dim_{data_source}")
```

```python
delta_table = DeltaTable.forName(spark, "fmcg.gold.dim_customers")
df_child_customers = spark.table("fmcg.gold.sb_dim_customers").select(F.col("customer_id").alias("customer_code"),"customer","market","platform","channel")
```

```python
delta_table.alias("target").merge(
    source=df_child_customers.alias("source"),
    condition="target.customer_code = source.customer_code"
).whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()
```
